```python
df=pd.read_csv('shots_data.csv')
```

```python
TeamA=df[df['team']=='Team A'].reset_index()
```

```python
TeamB=df[df['team']=='Team B'].reset_index()
```

```python
C3y_TeamA=TeamA[TeamA.y <= 7.8]
C3y_TeamA
```

```python
C3xp_TeamA=C3y_TeamA[(C3y_TeamA.x <= 23.75) & (C3y_TeamA.x >= 22.00)]
C3xp_TeamA
```

```python
C3xn_TeamA=C3y_TeamA[(C3y_TeamA.x >= -23.75) & (C3y_TeamA.x <= -22.00)]
C3xn_TeamA
```

```python
C3_TeamA= C3xn_TeamA.append([C3xp_TeamA])
C3_TeamA
```

```python
TwoPy_TeamA=TeamA[TeamA.y < 23.75]
TwoPy_TeamA
```

```python
TwoPx_TeamA=TwoPy_TeamA[(TwoPy_TeamA.x < 22.00) & (TwoPy_TeamA.x > -22.00)]
TwoPx_TeamA
```

```python
TwoP_TeamA=TwoPx_TeamA
```

```python
TwoP_TeamA
```

```python
NCR3_TeamA=pd.concat([TwoP_TeamA,C3_TeamA, TeamA]).drop_duplicates(keep=False)
```

```python
NCR3_TeamA
```

```python
Shots_Attempted_TwoPoint= len(TwoP_TeamA.index)/len(TeamA.index)*100
Shots_Attempted_TwoPoint
```

```python
Shots_Attempted_NonCorner3= len(NCR3_TeamA.index)/len(TeamA.index)*100
Shots_Attempted_NonCorner3
```

```python
Shots_Attempted_Corner3= len(C3_TeamA.index)/len(TeamA.index)*100
Shots_Attempted_Corner3
```

```python
Twopoint_made = sum(TwoP_TeamA['fgmade']==True)
Twopoint_made
```

```python
Corner3_made = sum(C3_TeamA['fgmade']==True)
Corner3_made
```

```python
NonCorner3_made = sum(NCR3_TeamA['fgmade']==True)
NonCorner3_made
```

```python
EFG_TwoPoint = ((Twopoint_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(TwoP_TeamA.index)*100
EFG_TwoPoint 
```

```python
EFG_Corner3 = ((Corner3_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(C3_TeamA.index)*100
EFG_Corner3 
```

```python
EFG_NonCorner3 = ((NonCorner3_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(NCR3_TeamA.index)*100
EFG_NonCorner3 
```
