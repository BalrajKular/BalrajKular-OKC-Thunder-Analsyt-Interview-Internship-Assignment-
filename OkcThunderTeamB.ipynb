```python
import pandas as pd
```

```python
df=pd.read_csv('shots_data.csv')
```

```python
df
```

```python
TeamB=df[df['team']=='Team B'].reset_index()
```

```python
TeamB
```

```python
C3y_TeamB=TeamB[TeamB.y <= 7.8]
C3y_TeamB
```

```python
C3xp_TeamB=C3y_TeamB[(C3y_TeamB.x <= 23.75) & (C3y_TeamB.x >= 22.00)]
C3xp_TeamB
```

```python
C3xn_TeamB=C3y_TeamB[(C3y_TeamB.x >= -23.75) & (C3y_TeamB.x <= -22.00)]
C3xn_TeamB
```

```python
C3_TeamB= C3xn_TeamB.append([C3xp_TeamB])
C3_TeamB
```

```python
TwoPy_TeamB=TeamB[TeamB.y < 23.75]
TwoPy_TeamB
```

```python
TwoPx_TeamB=TwoPy_TeamB[(TwoPy_TeamB.x < 22.00) & (TwoPy_TeamB.x > -22.00)]
TwoPx_TeamB
```

```python
TwoP_TeamB=TwoPx_TeamB
TwoP_TeamB
```

```python
NCR3_TeamB=pd.concat([TwoP_TeamB,C3_TeamB, TeamB]).drop_duplicates(keep=False)
NCR3_TeamB
```

```python
Shots_Attempted_TwoPoint= len(TwoP_TeamB.index)/len(TeamB.index)*100
Shots_Attempted_TwoPoint
```

```python
Shots_Attempted_NonCorner3= len(NCR3_TeamB.index)/len(TeamB.index)*100
Shots_Attempted_NonCorner3
```

```python
Shots_Attempted_Corner3= len(C3_TeamB.index)/len(TeamB.index)*100
Shots_Attempted_Corner3
```

```python
Twopoint_made = sum(TwoP_TeamB['fgmade']==True)
Twopoint_made
```

```python
Corner3_made = sum(C3_TeamB['fgmade']==True)
Corner3_made
```

```python
NonCorner3_made = sum(NCR3_TeamB['fgmade']==True)
NonCorner3_made
```

```python
EFG_TwoPoint = ((Twopoint_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(TwoP_TeamB.index)*100
EFG_TwoPoint 
```

```python
EFG_Corner3 = ((Corner3_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(C3_TeamB.index)*100
EFG_Corner3 
```

```python
EFG_NonCorner3 = ((NonCorner3_made)+(0.5*(NonCorner3_made+Corner3_made)))/len(NCR3_TeamB.index)*100
EFG_NonCorner3 
```
